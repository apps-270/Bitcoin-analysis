```python
# Step 1: Import libraries 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import warnings
warnings.filterwarnings('ignore')
sns.set(style='darkgrid')
```

```python
# Step 2: Load the data 
df = pd.read_csv(r"C:\Users\aparn\Downloads\btcusd_1-min_data.csv")
```

### Checking what columns we have in the dataset

```python
# Step 3: Get info about the data
df.info()
```

```python
# Step 4: Check for missing values 
df.isnull().sum()
```

```python
# Step 5: Convert Timestamp column to datetime format 
df['Timestamp'] = pd.to_datetime(df['Timestamp'])
df.set_index('Timestamp', inplace=True)
df.sort_index(inplace=True)
df.head()
```

### Let's draw the closing price of Bitcoin over time!

```python
# Step 6: Plotting closing price 
plt.figure(figsize=(14,6))
plt.plot(df['Close'], color='orange')
plt.title('Bitcoin Closing Price Over Time')
plt.xlabel('Year')
plt.ylabel('Price in USD')
plt.show()
```

### Now let's look at the trading volume

```python
print(df.columns)
```

```python
# Step 7: Plotting trading volume
import matplotlib.pyplot as plt

plt.figure(figsize=(14, 6))
plt.plot(df['Volume'], color='blue')
plt.title('Bitcoin Trading Volume Over Time')
plt.xlabel('Time')
plt.ylabel('Volume in USD')
plt.grid(True)
plt.show()
```

```python
# Step 8: Correlation heatmap
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Between Columns')
plt.show()
```
